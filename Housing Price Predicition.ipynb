```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("E:\ANACONDA\Housing.csv")
df
```

```python
df.info()
```

```python
obj_col = ['mainroad', 'guestroom', 'basement', 'hotwaterheating','airconditioning','prefarea', 'furnishingstatus'] 
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
df[obj_col] = df[obj_col].apply(lambda col: LE.fit_transform(col))
```

```python
df 
```

```python
x = df.drop('price', axis = 1)
x
```

```python
y = df['price'].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state = 0)
```

```python
print(x_train.size)
print(y_train.size)
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(x_train, y_train)
```

```python
y_test.size
```

```python
y_hat = LR.predict(x_test)
y_hat.size
```

```python
y_hat = LR.predict(x_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_hat.reshape(len(y_hat),1), y_test.reshape(len(y_test),1)),1))
```

```python
print(LR.intercept_)
print(LR.coef_)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_hat)
```
